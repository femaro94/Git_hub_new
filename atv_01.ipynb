Crie um programa em Python, que simule lançamentos de dados. 
Sorteie aleatóriamente um número dentre 1 e 6.

Depois sorteie novamente um número dentre 1 e 6 multiplique-os e armazene, obviamente, os resultados serão entre 1 e 36. 

I)  Crie um gráfico para visualizar a distribuição de números sorteados para apenas um Dado com sorteios de 1 a 6.

II)  Crie um gráfico para comparação do lançamentos dos dois dados de 1 a 6.

III)  Crie por fim, um gráfico demonstrando a distribuição dos lançamentos do produto dos resultados dos sorteio dos dois dados.

IV) Crie por fim, um gráfico demonstrando a distribuição dos lançamentos da soma dos resultados dos sorteio dos dois dados.

Faça 3 simulações:<br>

100 lançamentos<br>
1.000 lançamentos<br>
10.000 lançamentos

```python
import random
import matplotlib.pyplot as plt
```

```python
lancamento = 1000
```

```python
intervalo_multiplicacao = [c * i for c in range(1, 7) for i in range(1, 7)] #pesquei chatGPT
intervalo_multiplicacao = list(set(intervalo_multiplicacao))
intervalo_multiplicacao.sort()
print(intervalo_multiplicacao)
print(len(intervalo_multiplicacao,))
```

```python
dado_1 = [random.randint(1,6) for c in range(lancamento)]
dado_2 = [random.randint(1,6) for c in range(lancamento)]
multi_dado = [dado_1[c]*dado_2[c] for c in range(lancamento)]
soma_dado = [dado_1[c]+dado_2[c] for c in range(lancamento)]

intervalo_1 =  [{c:dado_1.count(c)} for c in range(1,7)] #gráfico de 1 a 6 com o primeiro dado
intervalo_2 = [{c:dado_2.count(c)} for c in range(1,7)] #gráfico de 1 a 6 com o segundo dado
intervalo_3 = [{c:multi_dado.count(c)} for c in intervalo_multiplicacao] #gráfico com os valores dos possiveis numeros de 1 a 6 multiplicado com a multiplicacao do primeiro com o segundo dado
intervalo_4 = [{c:multi_dado.count(c)} for c in range(2,13)] #gráfico com os valores de 2 a 12 com asoma do primeiro com o segundo dado

print(dado_1)
print(dado_2)
print(multi_dado)
print(soma_dado)
print(intervalo_1)
print(intervalo_2)
print(intervalo_3)
print(intervalo_4)
```

```python
# Grafico dado_1
intervalo = intervalo_1

# Extrair os valores de X e Y
x = [list(i.keys())[0] for i in intervalo]  # Pegando as chaves como X
y = [list(i.values())[0] for i in intervalo]  # Pegando os valores como Y

# Criar o gráfico de barras
plt.bar(x, y)
plt.xlabel("Face")
plt.ylabel("Frequência")
plt.title("Gráfico Lançamento do Dado1: 100")
plt.ylim(0, max(y) + max(y)*0.1)  # Ajusta o eixo Y para começar em zero até um pouco acima do valor máximo de y
plt.grid(True, axis='y')  # Grade apenas no eixo Y para visualização mais clara

# Adicionar rótulos aos pontos
for i in range(len(x)):
    plt.text(x[i], y[i], f"{y[i]}", ha='center', va='bottom')  # Centralizar rótulo em cima da barra

plt.show()
```
